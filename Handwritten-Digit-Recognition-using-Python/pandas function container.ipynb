````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
x = np.array([1,5,3,4,2])
y = np.array([5,7,9,11,13])
x_new=np.array([3,4,4,5,6,7,4,5,6])
````

````python
def onetwo(x):
  t=[]
  for i in x:
    t.append([i])
  return t
````

````python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(onetwo(x),y)
````

````python
y_predict=model.predict(onetwo(x_new))
````

````python
y_predict
````

````python
%matplotlib inline
def gradient_descent(x,y):
    m_curr = b_curr = 0 
    rate = 0.005
    n = len(x)
    plt.scatter(x,y,linewidths=14,color='red')
    plt.xlabel('x axis')
    plt.ylabel('y axis')
    
    for i in range(100000):
        y_predicted = m_curr * x + b_curr    # y=m*x + b ( b is intercept , m is slope)
        
        # if i==9999:
        #   print ('slop ',m_curr, ' intercet ',b_curr, ' this is itrestion ' ,i)
        
        plt.plot(x,y_predicted,'g--',)
        # plt.plot(x_new,y_predict,'c.-')
        
        md = -(2/n)* sum(x*(y-y_predicted))   # m is derivated
        yd = -(2/n)* sum(y-y_predicted)       # y is derivated
        
        # print(md,yd)
        m_curr = m_curr - rate * md       # 
        b_curr = b_curr - rate * yd  
    print(m_curr,b_curr)   
````

````python
gradient_descent(x,y)
````
