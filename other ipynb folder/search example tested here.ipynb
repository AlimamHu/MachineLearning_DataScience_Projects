```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris_file=datasets.load_iris()
```

```python
X=iris_file.data[:,2:]          #  'petal length (cm)', 'petal width (cm)'
Y=(iris_file.target==2).astype(np.float)    # virginica  true values is 1 & false valuse is 0
```

```python
from sklearn.pipeline import Pipeline   # pipeline make step wise step to process certain task
from sklearn.preprocessing import StandardScaler  # standarscaler use for large large margine 
from sklearn.svm import LinearSVC
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
svm_clf=Pipeline([
                  ('stander',StandardScaler())
 ,                 ('svc',LinearSVC(C=1,loss='hinge'))
])
```

```python
svm_clf.fit(X,Y)
```

```python
svm_clf.score(X,Y)
```

```python
model =LinearSVC(C=1,loss='hinge')
```

```python
model.fit(X,Y)
```

```python
model.score(X,Y)
```

```python
polynomial_clf=Pipeline([
                         ('poly_feature', PolynomialFeatures(degree=13)),
                         ('stander',StandardScaler()),
                         ('linearsvm',LinearSVC(C=10,loss='hinge'))
])
```

```python
polynomial_clf.fit(X,Y)
```

```python
polynomial_clf.score(X,Y)
```

```python
poly_x=PolynomialFeatures(degree=3)
```

```python
model.fit(poly_x.fit_transform(X),Y)
```

```python
model.score(poly_x.fit_transform(X),Y)
```
