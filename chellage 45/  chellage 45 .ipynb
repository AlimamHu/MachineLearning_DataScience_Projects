# DATA Preprocessing function 
take data and return X_data , y_data

```python
def preprocess_data(df_values,test_data=False):
    
    df_train=df_values.fillna(df_values.mean())  # fill na values to mean

    df_train['CustomerID']=df_train['CustomerID'].apply(lambda x:x.replace('CusID_','')).astype(np.int64)

    df_train=df_train.apply(lambda x:x.replace('?',df_train.median()))
    df_train=df_train.dropna(axis=0)

    # split date , month , year
    array_Sign_up_date=pd.to_datetime(df_train['Sign_up_date'],)
    df_train['Sign_up_day']=array_Sign_up_date.apply(lambda x:x.day)
    df_train['Sign_up_month']=array_Sign_up_date.apply(lambda x:x.month)
    df_train['Sign_up_year']=array_Sign_up_date.apply(lambda x:x.year)

    array_Last_order_placed_date=pd.to_datetime(df_train['Last_order_placed_date'],)
    df_train['Last_order_placed_day']=array_Last_order_placed_date.apply(lambda x:x.day)
    df_train['Last_order_placed_month']=array_Last_order_placed_date.apply(lambda x:x.month)
    df_train['Last_order_placed_year']=array_Last_order_placed_date.apply(lambda x:x.year)


    convert_col_int=['City','Gender','State']
    from sklearn.preprocessing import OneHotEncoder,LabelEncoder
    # enc = OneHotEncoder()
    enc=LabelEncoder()

    df_train['City']=enc.fit_transform(df_train['City'])
    df_train['Gender']=enc.fit_transform(df_train['Gender'])
    df_train['State']=enc.fit_transform(df_train['State'])


    if test_data:
        X_data=df_train.drop(['Sign_up_date','Last_order_placed_date',],axis=1)
        return X_data
    else:
        X_data=df_train.drop(['Sign_up_date','Last_order_placed_date','Preferred_Theme'],axis=1)
        y_data=df_train['Preferred_Theme']

        return X_data,y_data
    




    
    
```

```python
X_data ,y_data =preprocess_data(df_train)
```

# spliting data set into train & test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_data, y_data, test_size=0.10, random_state=42)  # X,y variable name
```

# Modeling

## Desision tree

```python
from sklearn.tree import DecisionTreeClassifier
deci_clf = DecisionTreeClassifier(random_state=3,max_depth=100)
```

## Naive byas

```python
from sklearn.naive_bayes import GaussianNB

gauss = GaussianNB()

```

## randomForest

```python
from sklearn.ensemble import RandomForestClassifier

ran_clf = RandomForestClassifier(max_depth=90, random_state=4,n_estimators=800) # increasing max_depth increasing training score

```

## SGDClassifier

```python
from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier(random_state=42,max_iter=1000)
```

## SVM

```python
from sklearn.svm import SVC
```

```python
sv_classifier=SVC(gamma='auto')
sv_classifier.fit(X_train,y_train)
```

# HELLO WORLD :)
so, we have three better model which have `low bias and low varience` , now i am **essamble model**

# Ensemble

```python
from sklearn.ensemble import VotingClassifier

model = VotingClassifier(estimators=[('sv', sv_classifier), ('sgd', sgd_clf),('ran',ran_clf),('dec',deci_clf)], voting='hard')
model.fit(X_train,y_train)

```

```python
print(f'training score --> {model.score(X_train,y_train)}')
print(f'test score --> {model.score(X_test,y_test)}\n','-'*15)
```

# cross validation score

```python
from sklearn.model_selection import cross_val_score

cross_val_score(model, X_train, y_train, cv=3, scoring="accuracy")
```
