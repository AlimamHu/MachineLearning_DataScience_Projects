```python
train_dataset_url='https://storage.googleapis.com/download.tensorflow.org/data/iris_training.csv'
```

```python
train_dataset_fp=tf.keras.utils.get_file(fname=os.path.basename(train_dataset_url),origin=train_dataset_url)
```

# this is minist data predict useing tensorflow

```python
model_data=tf.keras.datasets.mnist.load_data()
```

```python
(xtrain,ytrain),(xtest,ytest)=model_data
```

```python
x_train,x_test=xtrain/255.0,xtest/255.0
```

```python
model=tf.keras.Sequential([
                           tf.keras.layers.Flatten(input_shape=(28,28)),
                           tf.keras.layers.Dense(128,activation='relu'),
                           tf.keras.layers.Dropout(0.2),
                           tf.keras.layers.Dense(10)
])
```

```python
prediction=model(x_train[:1]).numpy()
```

```python
prediction
```

```python
tf.nn.softmax(prediction).numpy()*100
```

```python
loss_fuc=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
```

```python
loss_fuc(ytrain[:1],prediction)
```

```python
model.compile(optimizer='adam',
loss=loss_fuc,
metrics=['accuracy'])
```

```python
model.fit(xtrain,ytrain,epochs=5)
```

```python
model.evaluate(xtest,ytest,verbose=2)
```
