#### X TRAIN OR TEST

```python
df_diabetes.columns
```

```python
diabetes_X_train=df_diabetes[['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6']][:-20]
diabetes_X_test=df_diabetes[['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6']][-20:]
```

```python
print(len(diabetes_X_train))
print(len(diabetes_X_test))
```

Y TRAIN or TEST

```python
diabetes_Y_train=df_diabetes['target'][:-20]
diabetes_Y_test=df_diabetes['target'][-20:]
```

# model fitting area

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.get_params()
```

```python
model.fit(diabetes_X_train,diabetes_Y_train)
```

```python
y_estimate=model.predict(diabetes_X_test)
```

```python
model.score(diabetes_X_train,diabetes_Y_train)
```

```python
print(f'coef_ {model.coef_} \n\nintercept  {model.intercept_}')
```

```python
model.rank_
```

```python
model.singular_
```

```python
mean_squared_error(y_estimate,diabetes_Y_test)
```
